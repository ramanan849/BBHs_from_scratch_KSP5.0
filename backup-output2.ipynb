```python
# Part (5)
df = printCompasDetails(data2['BSE_Supernovae'])
count_int,count_list=0,0
for i in df:
    array = np.array(df[i])
    SN_state = array[29] #the item in the 29th index of the dataframe corresponding to each binary system is the "Supernova_State"
    #print(SN_state)
    if type(SN_state)==int:
        count_int+=1
    else:
        count_list+=1
print("Binaries with 0 Supernovae = {}\nBinaries with 1 Supernova = {}\nBinaries with 2 Supernovae = {}".format((10**4)-count_int-count_list,count_int,count_list))
```

```python
data3 = h5.File('Task-3/COMPAS_Output.h5','r')
```

```python
# Parts (1), (2) and (3)
event_history=getEventHistory(h5.File('Task-3/COMPAS_Output.h5','r'))
event_strings=getEventStrings(h5.File('Task-3/COMPAS_Output.h5','r'))
binary_no_interaction,total_stable_mt,total_cee=0,0,0
for j in event_history[1]:
    if len(j)==0:
        binary_no_interaction+=1
    if len(j)!=0:
        #print(j[0])
        if j[0][0]=='MT' and j[0][-1]==False and (j[0][-2]==True or j[0][-3]==True): # CEE = False and RLOF1/2 must be True
            total_stable_mt+=1
        if j[0][0]=='MT' and j[0][-1]==True: #CEE = True
            total_cee+=1
print('Total no. of binaries with no interation at all = {} out of 10,000\nFraction = {}'.format(binary_no_interaction,binary_no_interaction/(10**4)))
print('\nTotal no. of binaries that undergo CEE (Common Envelope Evolution) = {} out of 10,000\nFraction = {}'.format(total_cee,total_cee/(10**4)))
print('\nTotal no. of binaries having stable MT event (only) = {} out of 10,000\nFraction = {}'.format(total_stable_mt,total_stable_mt/(10**4)))
```

```python
# Part (4)
df_system_par=printCompasDetails(data3['BSE_System_Parameters'])
#list=[]
def find_mergers(df,num):
    mergers=0
    for i in (df.keys()[1:]):
        array=np.array(df[i])
        #mergers+=int(array[index])
        #print(array)
        if array[6]==num: #the parameter iin the 6th index position under seed in BSE_System_Paramaters is Evolution_Status
            mergers+=1
    return mergers
#print(list)
total_mergers=find_mergers(df_system_par,12) 
total_mergers_at_birth=find_mergers(df_system_par,13)
print("Total number of stellar mergers: {} out of 10,000\nFraction = {}".format(total_mergers,total_mergers/(10**4)))
print("\n*Total number of stellar mergers at birth: {} out of 10,000\nFraction = {}".format(total_mergers_at_birth,total_mergers_at_birth/(10**4)))
```
